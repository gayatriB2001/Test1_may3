```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\DELL\Desktop\ML\sales.csv',encoding = 'unicode_escape')
```

```python
df
```

```python
#1)drop unrelated columns
df.head()
df.drop('Status',axis=1,inplace=True)
df.drop('unnamed1',axis=1,inplace=True)
```

```python
#2)check for null values
df.isnull()
df.isnull().sum()
```

```python
#3 drop null values
df.dropna(axis=0,inplace=True)
df
```

```python
#4)what is the avaerage age of customer
df['Age'].mean()     # average of Age 
```

```python
#5)who are the majority buyers?(genderwise)
df.groupby('Gender')['Amount'].value_counts()
#df1= df.groupby('Gender')['Cust_name'].count()
```

```python
#6)where do our majority customers fall?(age)
x = df['Age Group']
plt.hist(x,bins=20)
plt.title('majority customers  fall')
plt.show()
```

```python
#7)which state has highest sales?
df2=df.groupby('State')['Amount'].value_counts()
df2

```

```python
# 8)How does occupation effect the purchase?
data = df.groupby('Occupation')['Orders'].mean().reset_index()
print(data)
plt.figure(figsize=(25,5))
plt.bar(data['Occupation'],data['Orders'])
plt.xlabel('Occupation')
plt.ylabel('Average Purchase')
plt.show()
```

```python
#9)which are the top 10 products be sold?


top = df['Product_Category'].value_counts().head(10).reset_index()
sold=top.columns = ['Product_Category','Amounts']
print(top)
plt.figure(figsize=(25,5))
sns.barplot(x =top['Product_Category'],y = top['Amounts'])
plt.xlabel('Product_Category')
plt.ylabel('Amounts')
plt.show()
```
